```python
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import tensorflow as tf





drive.mount('/content/gdrive')

# Note that the path to the file on Google Drive starts with /content/gdrive/MyDrive/ and then the file path relative to the root folder of your Google Drive. You can replace data.csv with the name of the file you want to read, and use other file reading functions like open
# or libraries like numpy to read other types of files.

file_path = "/content/gdrive/MyDrive/Data/VoyageCalc_8.csv"
data = pd.read_csv(file_path)

# You can now work with the data
#print(data.head())

# Load the data from a CSV file using Pandas
df = pd.read_csv(file_path)

# Define the index of the column to drop and all columns after it
cols_to_drop = df.columns[4:8].values
print('cols_to_drop')
print(cols_to_drop)

# Drop the columns using iloc indexing
df = df.drop(cols_to_drop, axis=1)


# Print the updated DataFrame
print(df.head())



data = df.values

# Normalize the data
data = (data - np.mean(data, axis=0)) / np.std(data, axis=0)

# Define the autoencoder
input_shape = (9,)
encoding_dim = 4

# Encoder
input_data = tf.keras.layers.Input(shape=input_shape)
encoded = tf.keras.layers.Dense(encoding_dim, activation='relu')(input_data)

# Decoder
decoded = tf.keras.layers.Dense(input_shape[0], activation='linear')(encoded)

# Autoencoder model
autoencoder = tf.keras.models.Model(inputs=input_data, outputs=decoded)

# Compile the autoencoder
autoencoder.compile(optimizer='adam', loss='mse')

# Train the autoencoder on the data
autoencoder.fit(data, data, epochs=50)

# Extract the encoded data
encoder = tf.keras.models.Model(inputs=input_data, outputs=encoded)
encoded_data = encoder.predict(data)

# Convert the encoded data to a DataFrame
encoded_df = pd.DataFrame(data=encoded_data, columns=['encoded_1', 'encoded_2', 'encoded_3', 'encoded_4'])

# Save the encoded data to a CSV file
encoded_df.to_csv('/content/gdrive/MyDrive/Encoded/encoded_data.csv', index=False) # replace 'encoded_data.csv' with the path to your output file



# Create the figure and the 3D axis
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Create the scatter plot
ax.scatter(encoded_df['encoded_1'], encoded_df['encoded_2'], encoded_df['encoded_3'], c=encoded_df['encoded_4'], cmap='viridis')

# Set the axis labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Show the plot
plt.show()



```

```python
!pip install pyproj
import pandas as pd
import pyproj

# Define the input GPS coordinates
latitudes = [37.7749, 37.7750, 37.7751, 37.7752]
longitudes = [-122.4194, -122.4193, -122.4192, -122.4191]

# Create a pandas DataFrame from the coordinates
df = pd.DataFrame({'latitude': latitudes, 'longitude': longitudes})

# Define the input and output coordinate reference systems
input_crs = pyproj.CRS('EPSG:4326') # WGS84 (GPS)
output_crs = pyproj.CRS('EPSG:3857') # Web Mercator

# Define the transformation function
transformer = pyproj.Transformer.from_crs(input_crs, output_crs, always_xy=True)

# Transform the GPS data to the output CRS
x, y = transformer.transform(df['longitude'], df['latitude'])

# Create a new DataFrame with the transformed coordinates
transformed_df = pd.DataFrame({'x': x, 'y': y})

# Add 50 feet to the x-coordinates
transformed_df['x'] += 50

print(df)
print(transformed_df)
```
